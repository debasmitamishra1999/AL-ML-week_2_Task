```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
read_csv=pd.read_csv("fifa.csv")
fifa=pd.DataFrame(read_csv)
print(fifa)
```

```python
fifa.head()
```

```python
fifa.isnull().sum()
```

```python
sns.heatmap(fifa.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Nationality',data=fifa)
```

```python
fifa.drop(['ID', 'Age', 'Overall', 'Value', 'Wage',	'Preferred Foot', 'International Reputation', 'Skill Moves', 'Position', 'Joined', 'Contract Valid Until', 'Height', 'Weight', 'Release Clause'],axis=1,inplace=True)
```

```python
fifa.head()
```

```python
fifa.shape
```

```python
club_counts = fifa["Club"].value_counts(normalize=True).count()
print(club_counts)
```

```python
Nationality_counts = fifa["Nationality"].value_counts(normalize=True).count()
print(Nationality_counts)
```

```python
potential_counts = fifa["Potential"].value_counts(normalize=True).count()
print(potential_counts)
```

```python
plt.figure(figsize=(15,32))
#Plot all the nations on Y Axisns.countplot(x='Club',data=fifa)
sns.countplot(y = fifa.Nationality,palette="Set2") 
```

```python
plt.figure(figsize=(14,7),dpi=651)
sns.countplot('Club',data=fifa.head(50),palette='CMRmap')
plt.xlabel('Club')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.title('Clubwise players in the Top 50',fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(14,7),dpi=164)
sns.countplot(x='Nationality',data=fifa.head(50),palette='CMRmap')
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.title('Country-wise players in the Top 50',fontsize = 20)
plt.show() 
```

```python
sns.set_style('whitegrid')
potential  = fifa.Potential
plt.figure(figsize = (12,8))
ax = sns.distplot(potential,bins = 60,kde = False,color ='red')
ax.set_ylabel('Number of Players')
ax.set_xlabel('Potential')
ax.set_title('Distribution of Potentials of Players')
plt.show()
```

```python
#checking for the top 15 playerss with highest potential scores  in the game
plt.figure(figsize=(10,5)) 
ax = fifa.groupby(['Name'])['Potential'].max().sort_values(ascending = False).head(15).plot.bar(color='0.1',  edgecolor='blue')
ax.set_xlabel('Player name')
ax.set_ylabel('Potential')
ax.set_title("Players with highest Potential")
plt.show()
```
